```python
import matplotlib.pyplot as plt

subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
marks = [97, 68, 59, 81, 77, 92]

plt.bar(subjects, marks, color='blue')
#plt.barh(marks, subjects, color='red')
plt.title("Bar Graph Example")
plt.xlabel ("Subjects")
plt.ylabel ("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
student1 = [97, 68, 59, 81, 77, 92]
student2 = [88, 61, 80, 40, 62, 52]
student3 = [54, 62, 77, 54, 71, 98]

index = np.arange(6)
width = 0.03

plt.bar(index, student1, width, color="red", label="Student 1")
plt.bar(index + width, student2, width, color="green", label="Student 2")
plt.bar(index + (width*2), student3, width, color="yellow", label="Student 3")

plt.title("Stacked Bar Graph Example")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.xticks(index + width/2, subjects)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_shares = [10, 40, 30, 5, 15]

Explode = [0, 0, 0, 0.5, 0]

plt.pie(market_shares, explode=Explode, labels=firms, shadow=True, startangle=45)


plt.axis('equal')
plt.legend(title = "List of Firms")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-15, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)

plt.show()
```

```python
import matplotlib.pyplot as plt

from numpy.random import rand

for color in ['red', 'green', 'yellow']:
    n = 100
    x, y = rand(2, n)
    scale = 450.0 * rand(n) ** 5
    plt.scatter(x, y, s=scale, c=color, alpha=0.3, edgecolors='blue')

plt.grid(True)
plt.show()
```
